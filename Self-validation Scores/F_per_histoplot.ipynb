```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import colors 

import scipy.io as sio
```

```python
valid_F_per_arr_content = sio.loadmat('valid_F_per_arr.mat')
valid_F_per_arr_content
```

```python
valid_F_per_arr = valid_F_per_arr_content['scores_tot']
valid_F_per_arr
```

```python
colors = ['skyblue', 'tan', 'lime']

dataset_name = ['EDSR', 'HR', 'EnhancedNet']
plt.hist(valid_F_per_arr, bins = 40, histtype = 'bar', label = dataset_name)

plt.title('Histogram of F_perceptual score distribution - DIV_validation 100 dataset')
plt.legend(prop = {'size': 10})
plt.grid(True)
plt.show()
```
